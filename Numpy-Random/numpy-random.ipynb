# numpy.random.normal
## Description
Draws random samples from a normal (Gaussian) distribution.

```python
import numpy as np
# Used for plotting
import matplotlib.pyplot as plot

# Sets a nice style for our plots
plot.style.use('ggplot')

# Just sets the plot to be bigger.
plot.rcParams['figure.figsize'] = [10, 9]
```

```python
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 1000)
```

```python
abs(mu-np.mean(s))
```

```python
abs(sigma - np.std(s, ddof=1))
```

```python
count, bins, ignored = plot.hist(s, 30, density=True)
plot.plot(bins, 1/(sigma * np.sqrt( 2 * np.pi)) *
          np.exp( - (bins-mu)**2 / (2 * sigma ** 2)),
          linewidth=2, color='r')
plot.show()
```
